## 1. Calculate the Total Bill
A restaurant bill is ₹450. GST is 18% and SGST of 12% and other ₹20, discount 20%. Calculate the final bill amount.

```python
# Write your solution here
bill= int(input("enter bill: "))
gst_tax=0.18
sgst=0.12

other=20

discount=0.20

final_bill = bill + bill*gst_tax +bill*sgst + other - discount*bill

print(f"Your final bill with our discount is : {final_bill}")
```

## 5. Area of a Circle

```python
# Write your solution here
radius= int(input("Enter radius: "))

pi=3.14
area= pi * radius * radius
print(area)

```
